```python
#Other Guy's Data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('Phys3Lab1CapstoneData.csv')

for i in range(1,10):
  time = df[f'Time (s) Run #{i}']
  theta = df[f'Angle (rad) Run #{i}']
  dtheta = df[f'Angular Velocity (rad/s) Run #{i}']
  alpha = df[f'Angular Acceleration (rad/s²) Run #{i}']
  position = df[f'Position (m) Run #{i}']
  velocity = df[f'Velocity (m/s) Run #{i}']
  acceleration = df[f'Acceleration (m/s²) Run #{i}']
  plt.plot(time,theta,'r.',label='Theta v. Time')
  plt.plot(time,dtheta,'g.',label='DTheta v. Time')
  plt.plot(time,alpha,'b.',label='Alpha v. Time')
  plt.xlabel('Time (s)')
  plt.ylabel('Rads')
  plt.legend()
  plt.show();
  print('')
  plt.plot(time,position,'r.',label='Position v. Time')
  plt.plot(time,velocity,'g.',label='Velocity v. Time')
  plt.plot(time,acceleration,'b.',label='Acceleration v. Time')
  plt.xlabel('Time (s)')
  plt.ylabel('Rads')
  plt.legend()
  plt.show();
```

```python

df = pd.read_csv('Phys3Lab1_AK.csv')
time = df[f'Time (s) Run #{6}']
theta = df[f'Angle (rad) Run #{6}']
position = df[f'Position (m) Run #{6}']
log_theta = np.log(theta)
log_position = np.log(position)
plt.plot(time,log_theta,'r.',label='Theta v. Time')
plt.ylabel('Theta (rads)')
plt.xlabel('Time (s)')
plt.legend()
plt.show();
print('')
plt.plot(time,log_position,'r.',label='Position v. Time')
plt.ylabel('Position (m)')
plt.xlabel('Time (s)')
plt.legend()
plt.show();
```
